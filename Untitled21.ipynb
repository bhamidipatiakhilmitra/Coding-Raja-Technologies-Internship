```python
# Data Preprocessing: Data Cleaning
import warnings
import pandas as pd  # Import pandas for data manipulation
import matplotlib.pyplot as plt  # Import matplotlib for data visualization
import seaborn as sns  # Import seaborn for statistical data visualization
from sklearn.linear_model import LinearRegression  # Import LinearRegression model from scikit-learn
from sklearn.decomposition import PCA  # Import PCA (Principal Component Analysis) for dimensionality reduction

warnings.filterwarnings('ignore') #Hides all warnings that are generated during runtime.
# Load the data from a CSV file
df = pd.read_csv("coffee quality data.csv")  # Load the dataset into a DataFrame
df.info()  # Output basic information about the dataset (e.g., column names, non-null counts, types)

# Check for missing values in the dataset
df.isnull() #missing values in dataset
df.isnull().sum()  # Sum of missing values for each column in the dataset

# Drop rows with missing values
df.dropna(inplace=True)  # Remove rows that contain missing (NaN) values from the dataset

# Check summary statistics after dropping rows with missing values
df.describe()  # Get statistical summaries of numerical columns (e.g., mean, std, min, max, percentiles)

# EXPLORATORY DATA ANALYSIS (EDA)

# Scatter plot to visualize relationship between 'Altitude' and 'Total Cup Points'
plt.figure(figsize=(8, 6))  # Set the size of the plot
plt.scatter(df['Altitude'], df['Total Cup Points'], color='red', marker='+')  # Create a scatter plot with 'Altitude' on x-axis and 'Total Cup Points' on y-axis
plt.xlabel('Altitude')  # Label for x-axis
plt.ylabel('Total Cup Points')  # Label for y-axis
plt.title('Scatter Plot: Altitude vs Total Cup Points')  # Title of the plot
plt.show()  # Display the plot

# Histogram to visualize the distribution of numerical features in the dataset
df.hist(bins=15, figsize=(15, 8))  # Create histograms for all numerical columns, with 15 bins and specified figure size
plt.suptitle("Distribution of Features")  # Title for the overall set of histograms
plt.show()  # Display the histograms

# Pairplot to visualize relationships between different numerical features
sns.pairplot(df)  # Create pairwise plots for all numerical features
plt.suptitle('Pairplot of Features')  # Title for the pairplot
plt.show()  # Display the pairplot

# Histogram to visualize the distribution of the target variable 'Total Cup Points'
plt.figure(figsize=(8, 5))  # Set the size of the plot
sns.histplot(df['Total Cup Points'], bins=15, kde=True, color='blue')  # Create a histogram for 'Total Cup Points' with 15 bins and a Kernel Density Estimate (KDE) overlay
plt.title('Distribution of Total Cup Points')  # Title for the histogram
plt.xlabel('Total Cup Points')  # Label for x-axis
plt.ylabel('Frequency')  # Label for y-axis
plt.show()  # Display the histogram

# Heatmap to check correlation between numerical features and identify strong relationships
plt.figure(figsize=(12, 6))  # Set the size of the plot
corr_matrix = df.corr()  # Compute the correlation matrix between numerical features
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')  # Create a heatmap of the correlation matrix, with annotated values and a 'coolwarm' color map
plt.title('Feature Correlation Heatmap')  # Title for the heatmap
plt.show()  # Display the heatmap

# Boxplot to detect potential outliers in 'Total Cup Points'
plt.figure(figsize=(8, 5))  # Set the size of the plot
sns.boxplot(x=df['Total Cup Points'])  # Create a boxplot for 'Total Cup Points' to check for outliers
plt.title('Boxplot of Total Cup Points')  # Title for the boxplot
plt.show()  # Display the boxplot

# FEATURE SELECTION

# For simplicity, we use 'Altitude' as a feature to predict 'Total Cup Points'

# Define the feature (X) and target (y) variables
X = df[['Altitude']]  # Selecting 'Altitude' as the independent feature
y = df['Total Cup Points']  # Setting 'Total Cup Points' as the dependent variable (target)

# PRINCIPAL COMPONENT ANALYSIS(PCA) - If applicable
# Here, we perform PCA to reduce the dimensionality if you have multiple features
pca = PCA(n_components=1)  # Initialize PCA to reduce the features to a single principal component (if there were multiple features)
X_pca = pca.fit_transform(X)  # Apply PCA transformation to 'Altitude' (reducing it to a single component)

# Modeling: Linear Regression

# Define and train the Linear Regression model
reg = LinearRegression()  # Initialize the LinearRegression model
reg.fit(X, y)  # Train the model using the 'Altitude' feature (X) and 'Total Cup Points' target (y)

# Predicting with the model
predicted_value = reg.predict([[1200]])  # Make a prediction for a new value of 'Altitude' (1200)
print(f"Predicted Total Cup Points for Altitude 1200: {predicted_value[0]}")  # Output the predicted 'Total Cup Points' for altitude 1200

# Model coefficients and intercept
print(f"Model Coefficients: {reg.coef_}")  # Output the coefficient(s) of the linear model (slope)
print(f"Model Intercept: {reg.intercept_}")  # Output the intercept of the linear model

# Optional: Visualize the regression line
plt.figure(figsize=(8, 6))  # Set the size of the plot
plt.scatter(df['Altitude'], df['Total Cup Points'], color='red', marker='+')  # Plot the scatter plot again
plt.plot(df['Altitude'], reg.predict(X), color='blue', linewidth=2)  # Plot the regression line
plt.xlabel('Altitude')  # Label for x-axis
plt.ylabel('Total Cup Points')  # Label for y-axis
plt.title('Linear Regression: Altitude vs Total Cup Points')  # Title for the plot
plt.show()  # Display the plot with the regression line




'''Data Preprocessing (Cleaning):

Loading data

Checking for missing values

Dropping rows with missing values

Displaying summary statistics

Exploratory Data Analysis (EDA):

Visualizing relationships with scatter plots, histograms, and pair plots

Analyzing correlation with heatmaps

Identifying outliers using boxplots

Feature Selection:

Selecting the feature (Altitude) and target variable (Total Cup Points).

Principal Component Analysis (PCA) (Optional):

PCA is useful when you have multiple features, and you want to reduce the dimensionality. I added it in case you want to scale it to multiple features.

Modeling (Linear Regression):

Defining and fitting the regression model

Making predictions and displaying coefficients/intercept

Visualizing the regression line with the scatter plot.'''
```
